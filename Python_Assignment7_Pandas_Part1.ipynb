1) How-to-count-distance-to-the-previous-zero 

```python
import pandas as pd
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]}) 
y_list=[]
z_idx=0
c_idx=0
for z in df['X']:
    c_idx+=1
    if z==0:
        z_idx=c_idx
    y_list.append(c_idx-z_idx)
print(y_list)
df["Y"]=y_list
df
```
